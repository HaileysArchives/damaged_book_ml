```python
import gradio as gr

# Step 1: 책 상태를 입력 받는 함수
def check_book_status(status):
    # 사용자로부터 받은 책 상태 정보 처리
    return f"선택한 상태: {status}\n다음은 사진 가이드라인입니다. 아래 지침을 참고하여 책의 상태를 사진으로 업로드 해주세요."

# Step 3: 사진 업로드 후 처리 함수
def upload_images(images):
    if images is not None:
        return f"{len(images)}장의 사진을 업로드 했습니다."
    else:
        return "사진을 업로드 해주세요."

# Step 4: YOLO 및 Gemini Vision 결과를 받아 처리 (여기선 가상 결과)
def display_results(images):
    # 가상 처리된 결과 텍스트
    example_result_text = "YOLO 모델과 Gemini Vision API 결과: 손상이 발견되었습니다."
    # Placeholder 이미지 URL (실제 YOLO나 API의 결과 이미지를 넣을 수 있음)
    example_image = "https://via.placeholder.com/300.png"
    return example_result_text, example_image

# UI 구성
with gr.Blocks() as demo:

    # Step 0: 사용자에게 질문
    chatbot_response = gr.Textbox(label="Chatbot", value="How can I help you?", interactive=False)

    # Step 1: 책 상태 선택
    with gr.Row():
        gr.Markdown("### Step 1: 책 상태를 선택하세요.")
        book_status = gr.Radio(label="책 상태", choices=["Ripped", "Wet", "Wear", "Stain", "Folded"])
        submit_status = gr.Button("상태 전송")

    # Step 2: 서버로부터 받은 메시지 (사진 가이드라인 표시)
    with gr.Row():
        guideline_message = gr.Textbox(label="사진 가이드라인", interactive=False)

    # Step 3: 사진 업로드 인터페이스
    with gr.Row():
        gr.Markdown("### Step 3: 가이드라인에 맞춰 책 사진을 4장 업로드하세요.")
        image_upload = gr.File(label="사진 업로드", file_types=["image"], file_count="multiple")
        submit_images = gr.Button("사진 업로드")

    # Step 4: 결과를 표시하는 섹션 (텍스트 + 이미지)
    with gr.Row():
        gr.Markdown("### Step 4: 결과 보기")
        result_text = gr.Textbox(label="결과 설명", interactive=False)
        result_image = gr.Image(label="처리된 이미지 결과")

    # Step 1: 상태 체크 및 가이드라인 표시
    submit_status.click(check_book_status, inputs=book_status, outputs=guideline_message)

    # Step 3: 이미지 업로드 처리
    submit_images.click(upload_images, inputs=image_upload, outputs=chatbot_response)

    # Step 4: YOLO 및 Gemini Vision 결과 표시 (텍스트와 이미지 결과)
    submit_images.click(display_results, inputs=image_upload, outputs=[result_text, result_image])

demo.launch()
```
