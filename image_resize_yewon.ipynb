드라이브에 리사이즈된 이미지 저장하기

```python
# 원본 이미지 폴더 경로 설정 (Google 드라이브에 있는 폴더)
image_folder = '/content/drive/MyDrive/Google_ML_Camp/wet_book'

# 리사이즈할 이미지 크기 설정 (640x640)
resize_to = (1080, 1080)

# 폴더 내 모든 파일 확인
image_files = os.listdir(image_folder)
print("Image files in the folder:", image_files)

# 리사이즈 함수 정의
def resize_images_in_folder(input_folder, output_folder, size=(1080, 1080)):

    # 출력 폴더가 없으면 생성
    os.makedirs(output_folder, exist_ok=True)

    # 입력 폴더의 모든 파일 확인
    image_files = os.listdir(input_folder)

    # 이미지 파일들 차례대로 리사이즈하고 저장
    for image_file in image_files:
        # 이미지 파일 경로 설정
        input_path = os.path.join(input_folder, image_file)

        # 이미지 열기
        try:
            with Image.open(input_path) as img:
                # 이미지 리사이즈
                img_resized = img.resize(size)

                # 리사이즈된 이미지를 저장할 경로 설정 (원본 파일 이름을 그대로 사용)
                output_path = os.path.join(output_folder, image_file)

                # 리사이즈된 이미지 저장
                img_resized.save(output_path)

                # 저장된 파일 경로 출력
                print(f"Resized image saved to {output_path}")
        except Exception as e:
            print(f"Error processing {input_path}: {e}")

# 이미지 리사이즈 함수 호출
input_folder = '/content/drive/MyDrive/Google_ML_Camp/wet_book'  # 원본 이미지 폴더 경로
output_folder = '/content/drive/MyDrive/Google_ML_Camp/resized_wet_book_1080'  # 리사이즈된 이미지 저장할 폴더 경로

# 원하는 크기로 리사이즈 (640x640 크기로)
resize_images_in_folder(input_folder, output_folder, size=(1080, 1080))
```
